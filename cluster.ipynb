```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000, n_features=3, centers=3, cluster_std=1)
```

```python
plt.scatter(X[:,0],X[:,1],alpha=0.5,s=2)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
```

```python
km.fit(X)
```

```python
km.score(X)
```

```python
plt.scatter(X[:,0], X[:,1],s=10, c=km.predict(X))
```
